### Part 1: TD Prediction: State Values

In this section, you will write your own implementation of TD prediction (for estimating the state-value function).

We will begin by investigating a policy where the agent moves:
- `RIGHT` in states `0` through `10`, inclusive,  
- `DOWN` in states `11`, `23`, and `35`, and
- `UP` in states `12` through `22`, inclusive, states `24` through `34`, inclusive, and state `36`.

The policy is specified and printed below.  Note that states where the agent does not choose an action have been marked with `-1`.

````python
policy = np.hstack([1*np.ones(11), 2, 0, np.zeros(10), 2, 0, np.zeros(10), 2, 0, -1*np.ones(11)])
print("\nPolicy (UP = 0, RIGHT = 1, DOWN = 2, LEFT = 3, N/A = -1):")
print(policy.reshape(4,12))
````

The above figure is what you will try to approximate through the TD prediction algorithm.

Your algorithm for TD prediction has five arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `policy`: This is a 1D numpy array with `policy.shape` equal to the number of states (`env.nS`).  `policy[s]` returns the action that the agent chooses when in state `s`.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `V`: This is a dictionary where `V[s]` is the estimated value of state `s`.

Please complete the function in the code cell below.

````python
def plotter(scores, num_episodes, plot_every=100):
    '''
    Plots the scores
    '''
    # plot performance
    plt.plot(np.linspace(0,num_episodes,len(scores),endpoint=False),np.asarray(scores))
    plt.xlabel('Episode Number')
    plt.ylabel('Average Reward (Over Next %d Episodes)' % plot_every)
    plt.show()
    # print best performance per X-episodes equal to plot_every number
    print(('Best Average Reward over %d Episodes: ' % plot_every), np.max(scores))
````

````python
def td_prediction(env, num_episodes, policy, alpha, gamma=1.0, plot_every=100):
    '''
    Creates a value dictionary of all the board spaces the agent encounters.
    '''
    
    # initialize empty dictionaries of floats
    V = defaultdict(float)
    
    # initialize performance monitor
    tmp_scores = deque(maxlen=plot_every)
    scores = deque(maxlen=num_episodes)
    
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODONE: complete the function
        
        # Initialize score
        score = 0
        
        # Observe State 0
        state = env.reset()
        
        while True:
            
            # Choose Action @ t using policy
            action = policy[state]
            
            # Take Action @ t and observe R[t+1] and S[t+1]
            next_state, reward, done, info = env.step(action)
            
            # Add reward
            score += reward
            
            # TD(0) formula
            V[state] = V[state] + alpha * (reward + gamma * V[next_state] - V[state])
        
            state = next_state
            
            # Check for terminal state
            if done:
                # Append score
                tmp_scores.append(score)
                break
        
        # Incrementally update the final score
        if (i_episode % plot_every == 0):
            scores.append(np.mean(tmp_scores))
    
    # Plot the results
    plotter(scores, num_episodes, plot_every)
    
    return V 
````

However, if you'd like to ensure the accuracy of the unit test, please do not change the value of `gamma` from the default.

````python
# evaluate the policy and reshape the state-value function 5000
V_pred = td_prediction(env, 50000, policy, .01)

# please do not change the code below this line
V_pred_plot = np.reshape([V_pred[key] if key in V_pred else 0 for key in np.arange(48)], (4,12)) 
check_test.run_check('td_prediction_check', V_pred_plot)
plot_values(V_pred_plot)
````

### Part 2: TD Control: Sarsa

In this section, you will write your own implementation of the Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def choose_action(env, Q, epsilon, state):
    '''
    Uses an epsilon greedy formula to choose the action
    '''
    nA = env.nA
    # Checks if state has already come up; if so, creates probabilities using the epsilon greedy method
    if state in Q:
        probs = np.ones(nA) * (epsilon / nA)
        best_a = np.argmax(Q[state])
        probs[best_a] = 1 - epsilon + (epsilon / nA)

    # If it's a new state, it uses even probability.
    else:
        probs = np.full(nA, 1 / nA)

    # Choose A[0] using e-greedy policy        
    action = np.random.choice(np.arange(nA), p=probs)
    
    return action
````

````python
def sarsa(env, num_episodes, alpha, gamma=1.0, plot_every=100):
    '''
    Using the Sarsa(0) method, creates an action-value function without a stated policy
    '''
    
    nA = env.nA
    
    # initialize action-value function (empty dictionary of arrays)
    Q = defaultdict(lambda: np.zeros(env.nA))
    
    # initialize performance monitor
    tmp_scores = deque(maxlen=plot_every)
    scores = deque(maxlen=num_episodes)
    
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()   
        
        ## TODONE: complete the function
        
        # Initialize score
        score = 0
        
        # Set epsilon to scale with the number of episodes
        epsilon = 1. / i_episode
        
        # Observe State 0
        state = env.reset()

        while True:
            # Set Action[t]
            action = choose_action(env, Q, epsilon, state)
            
            # Take Action @ t and observe R[t+1] and S[t+1]
            next_state, reward, done, info = env.step(action)

            # Add reward
            score += reward
            
            # Set Action[t+1]
            next_action = choose_action(env, Q, epsilon, next_state)
            
            # Sarsa(0) formula
            Q[state][action] = Q[state][action] + alpha * (reward + gamma * Q[next_state][next_action] - Q[state][action])
        
            state = next_state
            
            # Check for terminal state
            if done:
                # Append score
                tmp_scores.append(score)
                break
        
        # Incrementally update the final score
        if (i_episode % plot_every == 0):
            scores.append(np.mean(tmp_scores))
    
    # Plot the results
    plotter(scores, num_episodes, plot_every)
    
    return Q
````

However, if you'd like to ensure the accuracy of the unit test, please do not change the value of `gamma` from the default.

````python
# obtain the estimated optimal policy and corresponding action-value function
Q_sarsa = sarsa(env, 50000, .01)

# print the estimated optimal policy
policy_sarsa = np.array([np.argmax(Q_sarsa[key]) if key in Q_sarsa else -1 for key in np.arange(48)]).reshape(4,12)
check_test.run_check('td_control_check', policy_sarsa)
print("\nEstimated Optimal Policy (UP = 0, RIGHT = 1, DOWN = 2, LEFT = 3, N/A = -1):")
print(policy_sarsa)

# plot the estimated optimal state-value function
V_sarsa = ([np.max(Q_sarsa[key]) if key in Q_sarsa else 0 for key in np.arange(48)])
plot_values(V_sarsa)
````

### Part 3: TD Control: Q-learning

In this section, you will write your own implementation of the Q-learning control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def q_learning(env, num_episodes, alpha, gamma=1.0, plot_every=100):
    '''
    Using the Sarsamax/Q-learning method, creates an action-value function without a stated policy
    '''
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(env.nA))
    
    # initialize performance monitor
    tmp_scores = deque(maxlen=plot_every)
    scores = deque(maxlen=num_episodes)
    
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODONE: complete the function

        # Initialize score
        score = 0        

        # Set epsilon to scale with the number of episodes
        epsilon = 1. / i_episode
        
        # Observe State 0
        state = env.reset()

        while True:
            # Set Action[t]
            action = choose_action(env, Q, epsilon, state)
            
            # Take Action @ t and observe R[t+1] and S[t+1]
            next_state, reward, done, info = env.step(action)
            
            # Add reward
            score += reward
            
            # Sarsamax formula
            Q[state][action] = Q[state][action] + alpha * (reward + gamma * max(Q[next_state]) - Q[state][action])
        
            state = next_state
            
            # Check for terminal state
            if done:
                # Append score
                tmp_scores.append(score)
                break
        
        # Incrementally update the final score
        if (i_episode % plot_every == 0):
            scores.append(np.mean(tmp_scores))
    
    # Plot the results
    plotter(scores, num_episodes, plot_every)
    
    return Q
````

### Part 4: TD Control: Expected Sarsa

In this section, you will write your own implementation of the Expected Sarsa control algorithm.

Your algorithm has four arguments:
- `env`: This is an instance of an OpenAI Gym environment.
- `num_episodes`: This is the number of episodes that are generated through agent-environment interaction.
- `alpha`: This is the step-size parameter for the update step.
- `gamma`: This is the discount rate.  It must be a value between 0 and 1, inclusive (default value: `1`).

The algorithm returns as output:
- `Q`: This is a dictionary (of one-dimensional arrays) where `Q[s][a]` is the estimated action value corresponding to state `s` and action `a`.

Please complete the function in the code cell below.

(_Feel free to define additional functions to help you to organize your code._)

````python
def expected_sarsa(env, num_episodes, alpha, gamma=1.0, plot_every=100):
    '''
    Using the Expected Sarsa method, creates an action-value function without a stated policy
    '''
    
    nA = env.nA
    
    # initialize empty dictionary of arrays
    Q = defaultdict(lambda: np.zeros(nA))
    
    # initialize performance monitor
    tmp_scores = deque(maxlen=plot_every)
    scores = deque(maxlen=num_episodes)
    
    # loop over episodes
    for i_episode in range(1, num_episodes+1):
        # monitor progress
        if i_episode % 100 == 0:
            print("\rEpisode {}/{}".format(i_episode, num_episodes), end="")
            sys.stdout.flush()
        
        ## TODONE: complete the function
        
        # Initialize score
        score = 0
        
        # Set epsilon to a constant, since it doesn't work if it scales with the number of episodes
        epsilon = 0.0005 # 1. / i_episode
        
        # Observe State 0
        state = env.reset()

        while True:
            # Set Action[t]
            action = choose_action(env, Q, epsilon, state)
            
            # Take Action @ t and observe R[t+1] and S[t+1]
            next_state, reward, done, info = env.step(action)
            
            # Add reward
            score += reward
            
            # Expected Sarsa formula
            probs = np.ones(nA) * (epsilon / nA)
            probs[np.argmax(Q[next_state])] = 1 - epsilon + (epsilon / nA)
            expected = np.dot(Q[next_state], probs)
            Q[state][action] = Q[state][action] + (alpha * (reward + (gamma * expected) - Q[state][action]))
        
            state = next_state
            
            # Check for terminal state
            if done:
                # Append score
                tmp_scores.append(score)
                break
    
        # Incrementally update the final score
        if (i_episode % plot_every == 0):
            scores.append(np.mean(tmp_scores))
    
    # Plot the results
    plotter(scores, num_episodes, plot_every)
    
    return Q
````
